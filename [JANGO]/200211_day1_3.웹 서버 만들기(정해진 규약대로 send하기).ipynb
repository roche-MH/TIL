## 3.1. Hello 출력하기

```python
import socket

# 서버 소켓 만들기
server_socket = socket.socket(socket.AF_INET, # 통신 채널 선택, TCP/IP를 쓰겠다
                              socket.SOCK_STREAM)# TCP방식으로 만들겠다


# 연결
server_socket.bind(('localhost',  # 127.0.0.1과 같음.-> 자기port 자기가 리스닝
                    80))  # port 번호
# 데이터가 오는지 안오는지 확인
server_socket.listen(0) # 관리하는 port 개수(0:알아서 관리, )
print('lestening')

# 클라이언트가 접속
client_socket, addr = server_socket.accept()
print('accepting')

# 데이터를 64k로 나누어서 받음.
data = client_socket.recv(65535) # 최대 64k를 받을 수 있음.

print('receice data:', data.decode())

# 정해진 규약으로 보내기
# 헤더정보를 확인하고 body 정보 보내기
# \r: 내리고 \n: 앞으로 당기기 -> 타자기 생각
client_socket.send('HTTP/1.0 200 OK\r\n\r\nHello'.encode('utf-8'))


# 메시지 보내고 서버 종료
client_socket.close()
```

```python
import socket
from datetime import datetime
import subprocess

server_socket = socket.socket(socket.AF_INET, # 통신 채널 선택, TCP/IP를 쓰겠다
                              socket.SOCK_STREAM)# TCP방식으로 만들겠다

server_socket.bind(('localhost', 80))  
server_socket.listen(0) 
print('lestening')

# 한번만 실행하도록(코드를 바꾸지 않아도 됨!!!!꿀팁!!!)
while True:
    client_socket, addr = server_socket.accept()
    print('accepting')
    data = client_socket.recv(65535) # 최대 64k를 받을 수 있음.
    data = data.decode()
    filename = data.split('\r\n')[0].split(' ')[1]
    
    try:
        if 'py' in filename:
            # 표준 출력으로 꼭 보내야 함.
            # 웹에서 입력한 파라미터를 전송할 수 없음. 실행만 가능
            # 환경 변수 이용해서 파라미터를 실행파일로 보낼 수 있음.
            output = subprocess.check_output(['python.exe', '.' +filename])
            html = output.decode('cp949')
            client_socket.send( 'HTTP/1.0 200 OK\r\n\r\n'.encode('utf-8'))
            client_socket.send(html.encode('utf-8'))

        elif 'html' in filename:
            with open('.' + filename, 'rt', encoding = 'utf-8') as f: # text file
                html = f.read()
            client_socket.send( 'HTTP/1.0 200 OK\r\n\r\n'.encode('utf-8'))
            client_socket.send(html.encode('utf-8'))


        # favorite icon 여부 확인    
        elif '.jpg' in filename or '.ico' in filename:
            client_socket.send('HTTP/1.0 200 OK\r\n'.encode('utf-8'))
            client_socket.send('Content-Type: image/jpg\r\n'.encode())
            client_socket.send('Accept-Ranges: bytes\r\n\r\n'.encode())
            file = open('.'+filename, 'rb')
            client_socket.send(file.read())
            file.close()


        else:
            header = 'HTTP/1.0 200 OK\r\n\r\n'
            client_socket.send(header.encode('utf-8'))
    except Exception as err:
        print(err)
     
    client_socket.close()
```

##  함수로 정리

```python
import socket
import threading
from datetime import datetime
import subprocess

def httpprocess(client_socket) :
    data = client_socket.recv(65535)   
    data = data.decode()
    print(data)
    try :    
        headers = data.split("\r\n")
        filename = headers[0].split(" ")[1]
        
        if '.py' in filename:
            html = subprocess.check_output(['python.exe', filename.replace('/','')])
            html = html.decode('cp949')
            
            header = 'HTTP/1.0 200 OK\r\n'               
            client_socket.send(header.encode("utf-8"))
            client_socket.send("Content-Type: text/html\r\n\r\n".encode())                                    
            client_socket.send(html.encode("utf-8"))            
        elif '.html' in filename:
            file = open("."+ filename, 'rt', encoding='utf-8')
            html = file.read()    
            header = 'HTTP/1.0 200 OK\r\n\r\n'        
            client_socket.send(header.encode("utf-8"))
            client_socket.send(html.encode("utf-8"))
        elif '.jpg' in filename or '.ico' in filename:         
            client_socket.send('HTTP/1.1 200 OK\r\n'.encode())
            client_socket.send("Content-Type: image/jpg\r\n".encode())
            client_socket.send("Accept-Ranges: bytes\r\n\r\n".encode())
            file = open("." + filename, "rb")            
            client_socket.send(file.read())  
            file.close()               
        else :
            header = 'HTTP/1.0 404 File Not Found\r\n\r\n'        
            client_socket.send(header.encode("utf-8"))
    except Exception as e :
        print(e)         
    client_socket.close()


server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind(('localhost', 80))
server_socket.listen(0)
print("listening")

while  True :
    client_socket, addr = server_socket.accept()
    # 소켓 연산에서 시간제한 설정
    # 시간이 지나면 timeout 예외를 발생시킴.
    # 함수의 대기시간 -> clinet가 보내는 데이터가 없으면 err 발생
    # 내용도 없이 계속 요청이 들어오는 경우가 있기 때문에
    client_socket.settimeout(3)
    print("accepting")

    # 비동기식 모드: 함수 호출하고 바로 대기모드, 함수를 동시에 여러번 실행 가능
    t = threading.Thread(target=httpprocess, args=(client_socket,))
    t.start()
    
    # 동기식 모드: 함수가 모두 실행되고 난 후, 다음 함수 실행
#     httpprocess(client_socket) 
```
