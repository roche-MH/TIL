#  ORM
- 객페지향방식의 매핑방식의 DB
- 객체 지향 프로그래밍 언어를 사용하여 호환되지 않는 유형 시스템 간에 데이터를 변환 하는 프로그래밍

# 개요
## 1. 데이터, 정보
> 정보
- 관련성과 목적을 갖춘 데이터
- 정보: 데이터 + 의미

> 지식
- 데이터나 정보를 처리하는 방법이나 어떤 근거에 의한 판단을 내리는데 필요한 분석, 판단에 관한 법칙을 필요
- 내면화된 정보 + 정보 사용 능력

> 정보의 진화 단계
- 사실 또는 환경
- 데이터: 사실, 관찰
- 정보: 상황 설명을 위한 데이터=> 통계
- 지식: 의미있는 정보=> 전략/ 방법론
- 지혜: 통찰력을 가진 지식=> 철학, 문화

## 2. 데이터 베이스
- 일시적인 데이터: 해당 프로세스가 실행되는 동안만 존재(변수)
- 지속적인 데이터: 계속 존재, 비휘발성 매체에 저장
- 데이터 베이스: 지속적인 데이터의 모임(파일 시스템도 db의 일종)

#### 데이터베이스 관리 시스템
- 데이터베이스를 관리해주는 소프트웨어 시스템
- 예: oracle, My sgl, dbase, bada, infomix, access 등

#### 데이터 베이스 특징
> 데이터의 무결정
- 어떤 경로를 통해 들어왔든 어류가 있어서는 안됨.
- a라고 넣으면 a라고 존재, 변동되는 안됨.

>데이터의 독립성
- db와 응용프로그램은 서로 의존관계가 있으면 확장이 어려움

> 보안
- 데이터를 소유한 사람이나 접근이 허가된 사람만 접근가능
- 파일 시스템은 보안이 안됨(언제든 가져갈 수 있음.)
- 권한 설정 가능

> 데이터 중복 최소화
- 중복되어 저장하는 것을 방지

> 응용 프로그램 제작, 수정 용이
- 통일된 방식으로 응용프로그래믕ㄹ 작성할 수 있고, 유지 보수 쉬움

> 데이터의 안정성 향상
- 데이터가 손상되는 문제가 발생하더라고 원래의 상태를 복원, 복구할 수 있음.


## 3.DB 종류
#### 3.1.  계층형 DB
- xmflgudxo
- 1: N관계를 가짐
- 한번 구축하면 구조를 변경하기 어려움
- 접근의 유연성이 부족하여 임의 검색시 어려움
- dict와 같은 구조 

#### 3.2.관계형 DB
- 모든 데이터를 테이블에 저장
- 기본키와 외래키를 사용하여 관계를 맺음
- 업무 변화에 따라 바로 순응할 수 있고, 유지 보수도 편리
- 대용량 ㄷ이터를 체계적으로 관리 가능
- 데이터ㅢ 무결정 보장
- 시스템 자원을 많이 차지하여 시스템이 전반적으로 느려지는 단점.
- 두 개의 테이블이 어떤 관계를 갖고있는지 기술 가능 -> key 이용


#### 3.3 객체지향형 DB
- 관계형dB가 너무 많이 사용
- 더 효율적이지만..시장 점유 실패
